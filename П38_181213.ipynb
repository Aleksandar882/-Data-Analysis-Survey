```python
import pandas as pd
```

```python
data = pd.read_csv('/content/data-project-final.csv')
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
import plotly.express as px
```

```python
fig=px.bar(data, x='Age', color='Favorite',labels={
                     "Favorite": "Омилена социјална мрежа",
                     "Age": "Возрастна Група",
                     "count":"Број"
                 },
                title="Омилена социјална мрежа според возраст")
```

```python
fig.show()
```

```python
px.scatter_3d(data, x='Age', y='Favorite', z='Sex')
```

```python
fig2=px.bar(data, x='Favorite', facet_col='Eco_Status', labels={
                     "Favorite": "",
                     "Eco_Status": "Економски Статус",
                     "count":"Број"
                 },
                title="Омилена социјална мрежа според економски статус")
```

```python
fig2.update_layout(
    font_size=17
)
```
